## Ensemble——基因组数据库

* [Ensemble官网](https://www.ensembl.org/index.html?redirect=no)

* 与NCBI类似，但专注于基因组序列和注释。

* NCBI与Ensemble会进行每日**相互交换，实时汇总**。

* 我们在测序中，常发现这种类型的ID就是**Ensemble ID**
<img src="./picture/064.png" width="350"/>

<br>

* Ensemble主要由几大功能构成

  * [**Ensembl**](https://www.ensembl.org/index.html) ：提供基因组浏览、基因注释及序列查询功能
  * [**BioMart**](https://www.ensembl.org/biomart/martview)：批量查询基因组注释信息,可以用于批量id的转换，如[RNA的ID转换](https://blog.csdn.net/lotusng/article/details/88399218)
  * [**VEP**](https://www.ensembl.org/info/docs/tools/vep/index.html)：注释 DNA 或 RNA 序列上的变异，上传变异文件（VCF 格式）进行注释
  * [**Ensembl REST API**](https://github.com/Ensembl) ：提供编程接口，便于通过代码访问 Ensembl 数据
  * [**Ensembl BLAST/BLAT**](https://useast.ensembl.org/Multi/Tools/Blast)：根据序列相似性在Ensembl 数据库中查找相关的基因或蛋白质

<img src="./picture/072.png" width="1000"/>

* **冗余度**指的是数据中存在重复信息的程度。在生物信息学和数据库管理中，冗余度通常用于描述一个数据库中同一信息（如序列、记录）被多次存储的现象。<br>

| **冗余度类型** | **优点**                                                                 | **缺点**                                       |
|----------------|--------------------------------------------------------------------------|-----------------------------------------------|
| **高冗余度**   | - 数据完整，保留所有重复记录，便于细节研究。                             | - 数据量大，存储和计算成本高。                  |
|                | - 适合精确匹配和稀有变异的全面分析。                                      | - 分析效率低，可能导致处理时间增加。            |
|                | - 可追踪多种来源的序列及注释。                                           | - 容易产生信息冗杂，难以快速获取核心信息。      |
| **低冗余度**   | - 数据简洁，减少重复信息，便于快速分析和高效计算。                        | - 可能会丢失一些序列的细节或来源信息。          |
|                | - 降低存储和处理成本，支持大规模数据研究。                               | - 对特定的精确比对或溯源分析可能不够全面。      |
|                | - 提供代表性序列，适合进化分析和功能注释。                               | - 可能需要结合高冗余数据补充额外信息。          |

# 标识符Identifier转换工具

* **标识符（Identifier）** ： 是用于唯一标识某个生物学实体（如基因、蛋白质、物种、序列等）的一串字符或编号。标识符在数据库和工具之间传递数据时，起到了非常重要的作用，因为它能够确保不同的系统和数据库能够一致地引用同一对象。如：nsembl Gene ID（如 ENSG00000141510）

# 序列比较相关网站

## 如何评估序列相似？

* 序列相似有2个指标：**一致度** ，**相似度**

* **一致度**: 如采两个序列(蛋白质或核酸)长度相同，那么它们的一致度定义为他们对应位置上**相同**的残基占总长度的百分数。

* **相似度**: 如果两个序列(蛋白质或核酸)长度相同，那么它们的相似度定义为他们对应位置上**相似与相同**的残基的数目占总长度的百分数。

* **可为相识与相同的残基？**：
  **替换记分矩阵(Substitution Matrix)** ，反映残基之间相互替换率的替换记分矩阵，它描述了残基两两相似的量化关系。分为DNA替换记分矩阵和蛋白质替换记分矩阵。

  <img src="./picture/111.png" width="700"/>

## 常见的DNA替换记分矩阵

* **等价矩阵(unitarymatrix)** :最简单的替换计分矩阵，其中，相同的碱基之间匹配得分为1，而不同碱基间的替换得分为0。
  <br><br>由于这种矩阵不含有碱基的物理化学信息，而且不区别对待不同的替换，因此在实际研究中运用较少。

  <img src="./picture/112.png" width="350"/>

<br>

* **转换-颠换矩阵(transition-transversionmatrix)** :碱基的替换发生在嘧啶之间或嘌呤之间，则称为**转换**；而如果发生在嘧啶和嘌呤之间，则称为**颠换**<br><br>在进化过程中，转换发生的频率远比颠换高。为了反映这一情况，通常该矩阵中**转换**的得分为-1，而**颠换**的得分为-5

  <img src="./picture/113.png" width="350"/>

<br>

* **BLAST矩阵** :经过在研究过程中的大量实际对比发现，如果令被比对的两个核苷酸相同时得分为+5，反之为-4，则比对效果较好，这就是被广泛使用的BLAST矩阵。

  <img src="./picture/114.png" width="350"/>

***
<br>
<br>

## 常见的蛋白替换记分矩阵

* **等价矩阵(unitary matrix)** ：蛋白质等价矩阵和DNA等价矩阵原理相同，相同氨基酸的匹配得分为1，不同氨基酸的替换得分为0。实际研究中运用较少<br><br>

* **PAM矩阵** ：PAM矩阵是目前应用最广泛的计分方法之一，其基于进化原理。如果两种氨基酸替换频繁，说明自然界容易接受这种替换，那么这对氨基酸的替换得分就应该高。<br><br>**PAM-1**矩阵是进化产生的每一百个氨基酸发生一个突变的量值，**PAM-n**，即发生了n次突变<br><br>我们需要根据要比较的序列之间的亲缘关系远近来选择适合的PAM矩阵
    * 如果序列亲缘关系远，即序列间会有很多突变，就选择PAM-N中N数值较大的矩阵
    * 如果亲缘关系近，即序列间的突变较少，就选择PAM-N中N数值较小的矩阵。

<br>


* **BLOSUM矩阵(blocks substitution matrix)**
BLOSUM矩阵，也的是，有许多种矩阵带有编号。<br><br>
如**BLOSUM-90**，代表该矩阵是由一致度大于等于90%的序列计算而来的。<br><br>
BLOSUM都是通过对大量符合特定要求的序列计算而来的。
  * BLOSUM后编号较小的矩阵适合比较序列相似度低的序列（即亲缘关系远的序列）
  * BLOSUM后编号较大的矩阵适合比较序列相似度高的序列（即亲缘关系近的序列）

<br>

* **遗传密码矩阵GCM** :通过计算一个氨基酸残基转变到另一个氨基酸残基所需的密码子中碱基变化数目而得到的，矩阵元素的值对应付出于代价。

## 双序列比较——打点法

* **打点法** ：两个序列打点判断是否相似

<img src="./picture/117.png" width="700"/>

* 在打点矩阵中，**连续的对角线及对角线的平行线**代表两条序列中相同的区域。如图中的红线，蓝线，绿线

* 最后，我们放眼全局，**红色的线和蓝色的线加起来基本上构成了一条主对角线**。由此我们可以得出结论:序列1和序列2是比较相似的两条序列

<br>
<br>

<img src="./picture/118.png" width="600"/>

* 可以用一条序列自己对自己打点，寻找串联重复序列。这样的打点矩阵必然是对称的，并且有一条主对角线。

* 与**主对角线平行的短平行线**所对应的序列片段就是重复的部分。

<br>
<br>

<img src="./picture/119.png" width="600"/>

* 用这种方法我们还可以快捷的发现**短串联重复序列和重复次数**

* **短串联重复序列(short tandemrepeat,STR)** ：也叫做微卫星DNA.是一类广泛存在于真核生物基因组中的DNA串联重复序列。

* **STR**由2-6bp的核心序列组成，重复次数通常在15-30次。

* **STR**具有高度多态性即存在重复次数的个体间差异，而且这种差异在基因遗传过程中一般遵循孟德尔共显性遗传规律，所以它被广泛用于法医学个体识别、亲子鉴定等领域。

## 双序列比对alignment

* **序列比对法(alignment)** :两条序列的比对就是把 s 和 t 这两个字符串上下排列起来，在某些位置插入空格(gap)。

  然后依次比较它们在每一个位置上字符的匹配情况

  匹配的好，这个位置就会得高分

  匹配的不好，看看能不能左右错一错，或填上个空位，让附近的位置更好的匹配在一起，从而**使所有位置的得分之和尽可能的高**。

  说白了，就是通过插入空位，让上下两行中尽可能多的一致的和相似的字符对在一起

  需要使用专门的**序列比对算法**。分为**全局比对与局部比对**

## 多序列比较——BLAST

* 使用序列比对法(alignment)去比对数据库的每一条序列，耗时比较长。所以使用**BLAST算法**快速从数据库准确找到相似的序列

<br>

* **BLAST算法(Basic Local Alignment Search Tool)** 采取seeding-and-extending方法。 首先找到query序列和subject序列间匹配的seed， 然后沿种子向两边延伸，产生High Scoring Segement Pairs (HSPs)， 然后在这些特定区域运用Simith-waterman 算法，最后评估比对的identify。

***
<br>

* **BLAST的主要步骤**

  * **分词(Creat a word list)**
    >BLAST首先将输入的序列分解成固定长度的小片段。<br>  
    >**类比**：假设你的句子是“GATTACA”，BLAST会拆成这些小片段："GAT", "ATT", "TAC", "ACA"  


  * **种子匹配（Seed Matching）**
    >BLAST会把这些小片段和数据库中的索引(index)快速匹配，找到所有完全相同的词（**种子匹配**）。<br> <br> 
    >**类比**：这就像在图书馆中快速找到包含“GAT”或“ATT”等关键词的书。



  * **扩展匹配（Extension）**
    >一旦找到一个匹配的小片段，BLAST会从这个片段出发，尝试向两侧扩展匹配，计算相似性得分，直到得分下降到一定程度为止（**得分下降阈值**）。<br>  
    >**类比**：如果发现一本书有一句话和你的句子的一部分相同，就继续往前后查找，看整段话是否更相似。



  * **评分系统（Scoring System）**
    >BLAST会对每次比对结果进行打分  

  * **返回最佳结果（High-Scoring Pair, HSP）**
    >BLAST会对所有比对结果进行排序，返回得分最高的片段（**HSP**）。这些结果通常用 **E-value** 衡量（值越小，匹配越可靠）。

    >**E-value**表示在给定数据库规模下，随机出现某个比对得分或更高得分的**预期次数**(其不为概率，应大于1)。其公式如下：

    >$$
E = K \cdot m \cdot n \cdot e^{-\lambda S}
$$

    >$ E $：期望值，表示随机比对中出现得分不低于 $ S $ 的次数。<br>
    >$ K $：与搜索空间维度及特性相关的常数。<br>
    >$ m $：查询序列（query sequence）的长度。<br>
    >$ n $：数据库的大小。<br>
    >$ lambda $：与比对得分系统相关的比例参数。<br>
    >$ S $：比对的原始得分（根据得分矩阵计算）。<br>
    >$ e $：自然对数。

  * **序列比对(alignment)得出identify(一致度)**
   >BLAST会对所有（**HSP**)片段与query序列进行序列比对(alignment), 得出**identify(一致度)**

* **BLAST按算法分类**

  * **标准BLAST**

  * **PSI-BLAST(Position-Specific Iterated)**：基于标准 BLAST，利用初次比对的结果构建**位置特异性评分矩阵（PSSM）**，进行多次迭代比对，直到结果不再有显著变化时停止。

    用于寻找远缘的同源序列或者找蛋白家族新的成员。

    **位置特异性评分矩阵（PSSM）**： 是一种用于描述生物序列中每个位置上不同字符（如氨基酸或碱基）的频率和“权重”的矩阵
 
    在**PSI-BLAST**流程中，第一轮搜索会使用**标准的blast**，然后你需要选择第一轮搜索到的序列都将用来创建**PSSM**（默认使用全部序列创建**PSSM**）来进行**PSI-BLAST**<br> <br> 

    >假设 A 的朋 友只有 B，B 朋友除 A 外还有 C。
    >
    >如果输入序列的第一个位置是 A，那么在第一轮没有 PSSM 辅助的情况下，只有第一个位置是 A 或 B 的序列被找到了。它们是图 1-A 中所示的四条序列。
    >
    >根据这四条序列创建的 PSSM(图 1-B)得知，第一个位置可以是 A，也可以是 B，那么 在第二轮搜索中，除了 A 的朋友 B 之外，B 的朋友 C 也可以出现在第一个位置了。这样如此反复，我们就可以找到朋友的朋友了。
    
    ><img src="./picture/149.png" width="700"/>
    
    <br>
    
    
  * **PHI-BLAST(Pattern-Hit Initiated)** ：输入序列相似的并符合某种特定**模式(pattern)**, **模式(pattern)** 是对特征的描述, 特定模式可通过正则表达式来表述(正则表达式就是这句话的一个简约的规范性字符书写法)
 
    > 比如:发生 N 糖基化位点的序列都符合这样一个特定的模式：发生糖基化的天冬酰胺，后面一定紧跟一个脯氨酸以外的氨基酸，再紧跟丝氨酸或者苏氨酸，再紧跟一个脯氨酸以外的氨基酸
    >
    > 发生 N 糖基化位点的序列符合的特定模式翻译成正则表达式为 **N{P}[ST]{P}** 。
    >
    > 其中，**N**是天冬酰胺，**P**是脯氨酸，**S**是丝氨酸，**T**是苏氨酸<br>
    > **{}** 代表除大括号里的氨基酸以外的任意氨基酸， **[ ]** 代表中括号中的任意一个氨基酸
    
    <br>

    > 在正则表达式中，除了以上，还有：**x** 代表任意字母， **(3,7)** 代表 3 到 7 个某字符。
    > 
    > **{L}GEx[GAS][LIVM]x(3,7)**
    > 
    >  除**L**以外的任意一个字母，紧接**G**，再紧接**E**，再接一个任意字符，之后是**GAS**中的任意一个，再接**LIVM**中的任意一个，最后再接 3 到 7 个任意字符

## 多序列比对multiple alignment

* **多序列比对(multiple alignment)** :对两条以上的生物序列进行全局比对

<br>

* **多序列比对的主要用途**:

  * 确认:一个未知的序列是否属于某个家族
  
  * 建立:系统发生树，查看物种间或者序列间的关系

  * 模式识别:一些特别保守的序列片段往往对应重要的功能区域，通过多序列比对，可以找到这些保守片段

  * 已知推未知:把已知有特殊功能的序列片段通过多序列比对做成模型，然后根据该模型推测未知的序列片段是否也具有该功能。

  * 其他:预测蛋白质/RNA二级结构

<br>

* 目前所有的多序列比对工具都不是完美的，它们都使用一种近似的算法。

<img src="./picture/153.png" width="800"/>

<br>

* **对要进行多序列比对的序列的要求**:

  * **不能太多的序列**。一般10-15条序列，最好别超过50条

  * **关系太远的序列不行**。两两之间序列相似度低于30%的一组序列作多序列比对会有麻烦。

  * **关系太近的序列不行**。两两之间序列相似度大于90%的序列，有再多条都等于只有一条。

  * **短序列受不行**。多序列比对支持一组差不多长的序列，个别很短的序列属于捣乱分子。

  * **有重复域的序列不行**。如果序列里包含重复域大多数多序列比对的程序都会出错，甚至崩溃。
 
<br>

* **给序列起名字的要求**：

  * 序列名字里不要有“空格”，用下划线代替“空格”是个好习惯。

  * 不要使用特殊字符，比如中文，@，#之类的。

  * 序列名字不要太长，最好在 15 个字符以内。

  * 一组序列里，不要有重名的序列。

## Primer-BLAST——引物特异性验证

* [Primer-BLAST官网](#https://www.ncbi.nlm.nih.gov/tools/primer-blast/index.cgi?LINK_LOC=BlastHome)

* 当我们设计好引物时，要验证该引物是否与其他gene发生反应，其**特异性**是否好，这时候就需要用到blast工具了。


<img src="./picture/026.png" width="800"/>
<img src="./picture/027.png" width="800"/>

<br>
<br>

* 通过blast，我们往下拉可以看到比对出了很多序列。如果结果都是目标gene，且排在前面，则说明该引物的特异性非常好

* 我们选择引物的时候，要选择跟**别的基因转录本匹配差**的特异性引物。

<img src="./picture/028.png" width="800"/>

<br>

>**Self-complementarity**用于预测引物本身甚至引物对之间可能的二级结构。Self complementarity**0-3**为**理想**状态 ，其值**越低越好**。<br>
>**Self 3' complementarity** 特别关注3'末端的自我互补性，其值**严格**为＜引物自身**1/4碱基数**，其值**越低越好**。
